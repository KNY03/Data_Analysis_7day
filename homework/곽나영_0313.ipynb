(2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.

```python
import seaborn as sns

iris = sns.load_dataset("iris")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =  False # 마이너스 부호가 깨지는걸 막기 위해서
```

```python
df1 = iris.groupby("species")["sepal_length"].mean()
df1.plot(kind='bar', rot=0)
plt.title("species별 sepal_length 평균")
plt.xlabel("species")
plt.ylabel("평균 sepal_length")
plt.show()
```

```python
df2 = iris["species"].value_counts()
df2.plot.pie(autopct='%.2f%%')
plt.title("품종별 데이터 개수 비율")
plt.axis('equal')
plt.show()
```
