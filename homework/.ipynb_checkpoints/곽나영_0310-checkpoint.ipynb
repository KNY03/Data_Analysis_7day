[문제1] 다음의 Series를 생성하고 sample 변수에 대입하고 출력하는 코드를 작성하시오.

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50],     # 일반 매개변수
             index = ["가", "나", "다", "라", "마"])    # 키워드 매개변수 /  인텍스를 문자열로 지정을 할 수 있음
print(s)
```

[문제2] 아래의 코드를 사용하여 sample2 중 160 이하인 데이터만 필터링여 출력하는  코드를 작성하시오.

```python
pip install seaborn
```

```python
import numpy as np

np.random.seed(20)
sample2 = pd.Series(np.random.randint(100, 200, size=(15,)))

print(sample2[sample2 <= 160])

```

[문제3] 다음과 같은 Series를 생성하는 코드를 작성하시오.

```python
s = pd.Series({
    "가": "apple",
    "나": np.nan,
    "다": "banana",
    "라": "kiwi",
    "마": "gubong"
}) 

s
```

[문제4] 위 예제에서 결측치가 아닌 데이터만 필터링하여 출력하는 코드를 작성하시오.

```python
s = s[s.notnull()]
print(s)
```

[문제5] 위 예제에서 결측치 데이터만 필터링하여 출력하는 코드를 작성하시오.

```python
s = pd.Series({
    "가": "apple",
    "나": np.nan,
    "다": "banana",
    "라": "kiwi",
    "마": "gubong"
}) 

s = s[s.isna()]
print(s)
```

[문제6] 아래 코드를 사용하여 sample에서 다음과 같은 결과를 가지도록 슬라이싱하여 출력하는 코드를 작성하시오.

```python
np.random.seed(0)
sample = pd.Series(np.random.randint(100, 200, size=(10,)))


print(sample[2:7]) 
```

[문제7] 아래 코드를 사용하여 sample2에서 다음과 같은 결과를 가지도록 슬라이싱하여 출력하는 코드를 작성하시오.((1)~(3)번 세문제임)

```python
np.random.seed(0)
sample2 = pd.Series(np.random.randint(100, 200, size=(10,)), index=list('가나다라마바사아자차'))

print(sample2["바": "차"])
print(sample2["가": "다"])
print(sample2["나": "바"]) 
```
