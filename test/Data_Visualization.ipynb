### cumsum()이란?
누적합을 구하는 함수

sum() : 전체 합을 한 번에 구함

cumsum() : 더해지는 과정을 순서대로 보여줌


```python
df = pd.DataFrame({'col1': [1,-2,5,3,-1,7], 'col2': [3,4,np.nan,9,2,5]})
df
```

```python
df.cumsum(skipna=False)
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail(3)
```

```python
df1.plot()
plt.title("Pandas의 plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

```python
df1.plot(kind = 'line')
plt.title("Pandas의 plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name="feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0, 8)
plt.show()
```
