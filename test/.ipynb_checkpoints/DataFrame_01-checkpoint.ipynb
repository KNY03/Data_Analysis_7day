```python
import pandas as pd

period = pd.period_range(
    start='2022-01-13 00:00:00',
    end='2022-01-13 02:30:00',
    freq='30min'
)

data = {
    "col1": [1, 2, 3, 4, 5, 6],
    "col2": period
}

idx = ["row1", "row2", "row3", "row4", "row5", "row6"]

df = pd.DataFrame(data, index=idx)
df
```

```python
df.rolling(window=3).sum(numeric_only=True)  # rolling이란 어떤 기간을 주고 그 걸 두하는 방식
# window = 3이란 데이터 3개를 가지고 구해라 이기 때문에 앞에 두개는 인식이 안된다
# 그렇기 때문에 앞에 2 값은 데이터가 NaN으로 출력된다.
```

```python
df.rolling(window=3).mean(numeric_only=True) 
# 이러한 것들은 주식같은 데이터에서 사용될 수 있다
```

```python
df = pd.DataFrame({"A":[1,4,7], "B":[2,5,8], "C":[3,6,9]})
df
```

```python
df.iloc[0]  # iloc(), loc() 함수를 사용하면 행-렬 순서로 사용된다.
# 지금과 같이 행으로 나온다.
```

```python
df.iloc[[0,2]] # 이것도한 마찬가지로 행-렬 순서로 데이터를 출력한다.
```

```python
df.loc[0]  # 지금 위치 인덱스가 아니라 열 정수의 이름을 넣은 것이다. 한마디로 라벨명이다.
```

```python
df.loc[:,"A"]  # 이것또한 행-렬 순서이다. 꼭 기억해야함!!
```

```python
# 이 함수들을 사용하면 행-렬 함수다
df.loc[:,"A":"B"]  # 열이 2개이다 열이 1개이면 시리즈로 2개이면 데이터 프레임으로 자동으로 만들어진다
```

```python
df["A"] # 이렇게 직접 접근하면 열-행의 순서로 기술 해야한다. 
```

```python
df.loc[0]["B"]
```

```python
df["B"][0]
# 위 두개를 보면 같은 값을 참조한다.
```
