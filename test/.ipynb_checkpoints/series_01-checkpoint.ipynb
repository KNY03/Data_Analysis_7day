### Series 클래스 indexing
- 인덱스 방법 이외에도 인덱스 라벨을 이용한 인덱싱 가능
- 배열 인덱싱
- 슬라이싱(slicing)

```python
s1 = Series([10, 20, 30])
s2 = Series([10, 20, 30], index=[1, 2, 3])

print(s1[0])
# print(s2[0]) 이렇게 하면 오류가 남 / 인덱스 번호를 다르게 지어했기 때문
print(s2[1]) 
```

```python
s3 = Series([10, 20, 30], index=["a", "b", "c"])
print(s3["a"])
```

```python
print(s3.index[0])
print(s3.index[-1])
# 잘 사용하진 않지만 그래도 알고 있는게 좋을 듯
```

### 슬라이싱

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],    
             index = ["서울", "부산", "인천", "대구"]) 
s
```

### series dic(딕셔너리) 자료형

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],     
             index = ["서울", "부산", "인천", "대구"]) 
s
```

```python
print("대구" in s) # in 연산자는 안에 값이 있는지 없는지 확인 가능
```

```python
print("대전" in s)
```

```python
for key, value in s.items():
    print("%s : %d" % (key, value))
```

딕셔너리는 순서를 가지지 않기 때문에 시리즈 데이터 순서가 보장되지 않는다.

```python
s2 = pd.Series({"서울":9904312, "부산":3448737, "인천":2890451,  "대구":2466052})
print(s2)     # 데이터를 기술한 순서대로 출력된다
```

```python
s = pd.Series({"서울":9904312, "부산":3448737, "인천":2890451,  "대구":2466052},  
             index = ["부산", "서울", "인천", "대구"]) 
s  # 인덱스 배치 순서에 따라서 출력
```

```python
from pandas import Series

data = [1000, 2000, 3000]
index = ["메로나", "구구콘", "하겐다즈"] 

s = Series(data=data, index=index) 
s
```

```python
print(s.iloc[0:2]) # i가 붙으면 위치 인덱싱이고 없으면 라벨 인덱싱이다
print(s.loc["메로나":"구구콘"])
```

```python
indic = [0,2]
print(s.iloc[indic]) # 추출할 위치 인덱스 번호를 리스트로 지정하여 해당 행을 가져옴
```

```python
print(s.iloc[[0,2]])
```

```python
indic = ["메로나",  "하겐다즈"] 
print(s.loc[indic])
print(s.loc[["메로나",  "하겐다즈"]])
```

```python
s # 현재 상태 확인
```

```python
s.loc["메로나"]  = 500  # 데이터 수정
# 또 다른 방법으로는 s.iloc[0] = 500 / s['메로나'] = 500
print(s)
```

```python
s.loc["비비빅"] = 5050 # 없는 값 이기때문에 행 인덱스를 추가한다
print(s)
```

```python
# 삭제할때 drop 을 사용함
# 만약 기존 객체에 삭제된 데이터를 저장하고 싶으면 (수정된) 그러면 저장을 해주야 한다
s1 = s.drop("메로나")
print(s)
print(s1)
```

### 인덱스 기반 연산

자료가 존재하지 않아 연산이 불가능한 경우 Nan으로 표시

float 자료형에서만 가능하므로 Nan이 다른 계산 결과도 Nan이다

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],    
             index = ["서울", "부산", "인천", "대구"]) 
s2 = pd.Series({"서울":48561, "부산":58951, "인천":78462,  "대전":256845},  
             index = ["부산", "서울", "인천", "대전"]) 

ds = s - s2    # 어떤 값과 Nan을 연산하면 Nan이다. 그리고 Nan과 계산이 되면 타입은 float이다
print(ds)
```

```python
s3 = s / 100000 # 각 데이터에 각각 연산이됨 / 이러한 연산을 브로드케스팅 연산이라고 함
print(s3)
```

```python
print(s.values - s2.values)
```

```python
print(ds[ds.notnull()])  # Null값을 포함 하지 않겠다 라는 의미(밑에서 볼 수 있듯이 참인 값만 출력한다)
```

```python
ds.notnull() # 이건 Bool값으로 반환되는데 True인 값만 위에서 출력한다
```

```python
rs = (s - s2) / s2 * 100 # 일치하지 않는 키는 Nan으로 처리됨
print(rs)
```

```python
rs = rs[rs.notnull()]
print(rs)
```

```python
rs["부산"] = 1.63  # 수정
rs["대구"] = 1.41  # 추가
print(rs)
```

```python
del rs["서울"] # drop과 같이 삭제의 의미
print(rs)
```

```python
from pandas import Series

date = ["6/1", "6/2", "6/3", "6/4", "6/5"]   # 동일한 라벨이기 때문에 연산이 가능하다. 
high = Series([42800, 42700, 42600, 42500, 43000], index=date)
low = Series([42150, 42315, 42300, 41300, 52340], index=date)
print(high)
print("-------------------")
print(low)
```

```python
diff = high - low # 동일한 라벨이기 때문에 연산이 가능하다. 
print(diff)
```

```python
print(diff.max()) # 가장 큰 값을 알수있다
```

```python
print(diff.min()) # 가장 작은 값을 알수있다
```

```python
print(diff.idxmax())  # 가장 큰 값의 위치를 알수있다
```

```python
def remove_comma(x):
    print(x, 'in function')
    return x
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = s.map(remove_comma)   #  시리즈 객체가 가지고 있는 것 만큼 반복한다
result = list(s.map(remove_comma)) # 리스트 형태로도 반환이 가능하다
print(result)
```

```python
def remove_comma(x):
    return int(x.replace(",", "")) # replace는 처음값을 다음 값으로 치완해라 라고 하는 함수이다
```

```python
result = list(s.map(remove_comma)) # 리스트 형태로도 반환이 가능하다
print(result)
```

## 데이터 정렬

```python
from pandas import Series

data = [3.1, 3.2, 2.0, 10.1, 5.1]
index = ["000010", "000020", "000030", "000040", "000050"]
s = Series(data, index)
print(s)
```

```python
s1 = s.sort_values() # 값을 기준으로 오름차순 정렬 문자는 코드값으로 정렬된다
print(s1)
```

```python
s1 = s.sort_values(ascending=False) # 값을 기준으로 내림차순 정렬 문자는 코드값으로 정렬된다
print(s1)
```

```python
s2 = s.rank() # 순위를 매긴다. 어떤 순위일까 - 인덱스의 순위에 따라서 오름차순
print(s2)
```

```python
s2 = s.rank(ascending=False) # 순위를 매긴다. 어떤 순위일까 - 인덱스의 순위에 따라서 내림차순
print(s2)
```
